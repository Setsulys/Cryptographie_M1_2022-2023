```python
def F(B,C,D):
    return [D[i] ^( B[i] and (C[i] ^ D[i])) if i < 16 else C[i] ^( D[i] and (B[i] ^ C[i])) for i in range(32)]
```

```python
def encode_byte_by_byte(message):
    lst=list()
    bytes_message = bytes(message,'UTF-8')
    for byt in bytes_message:
        lst.append(bytes(byt))
    return lst
```

```python
encode_byte_by_byte("hackerman")
```

```python
def preprocessing(message):
    lst = list()
    for n in message:
        lst.append(n)
    lst.append(bytes(1))
    print(lst)
```
